```python
from sklearn.cluster import KMeans
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
```

```python
with open('Animals_with_Attributes/predicate-matrix-continuous.txt') as f:
    featureData = np.array([[float(x) for x in line.strip().split()] for line in f.readlines()])
```

```python
with open('Animals_with_Attributes/classes.txt') as f:
    animals = [line.strip().split()[1] for line in f.readlines()]
```

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(featureData)
```

```python
clusters = [[] for i in xrange(0,10)]
for i in xrange(0,50):
    clusters[kmeans.labels_[i]].append(animals[i])
```

```python
clusters
```

```python
wardMethod = linkage(featureData, method = 'ward')
```

```python
plt.figure()
dendrogram(wardMethod, orientation = 'right', labels = animals)
rcParams['figure.figsize'] = 1,1
plt.show()
```
