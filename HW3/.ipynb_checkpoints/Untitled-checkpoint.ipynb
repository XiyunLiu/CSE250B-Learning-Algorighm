```python
class LinearRegressionUsingCoordinateDescent():
    def __init__(self):
        self.stepSize = 1e-6
        self.converge = 1e-5
        self.maxIter = 1e6
        self.coeffi = []
        self.numClass = 3
        self.trainingData = None
        self.trainingLabel = None
        self.numTrainingData = None
        self.loss = []
        self.wUpdatingList = []
        
    def lossFunction(self,w,trainData,trainLabel):
        lossList = map(lambda x: np.power((np.dot(x[:-1],w) - x[-1]),2), zip(trainData,trainLabel))
        loss = reduce(lambda x,y: x+y, lossList)
        return loss/2
    
    def fit(self, trainingData, trainingLabel, randomize = False):
        self.trainingData = trainingData
        self.trainingLabel = trainingLabel
        self.numTrainingData = len(trainingLabel)
        
        Y = -np.ones((3,len(trainingData)))
        for i in range(0,len(trainingLabel)):
            Y[int(trainingLabel[i])-1][i] = 1
            
        for curClass in range(self.numClass):
            if randomize:
                curCoeff = self.generateCoeffRandom(trainingData, Y[curClass])
            else:
                curCoeff = self.generateCoeffBatch(trainingData, Y[curClass])
            self.coeffi.append(curCoeff)
    
    def generateCoeffBatch(self, X,y):
        w = np.zeros(14)
        j = 0
        iteration = 0
        lossList = []
        wUpdatingPath = []
        lossChangeList = np.zeros(14)
        differentiateList = np.zeros(14)
        differentiate = np.dot(X.T[j], np.subtract(np.dot(X,w),y))
        postLoss = self.lossFunction(w,X,y)
        while iteration <= self.maxIter and (j == 13 or 0 in differentiateList or max(differentiateList) > self.converge):
            preLoss = postLoss
            for t in range(0,100):
                w[j] -= self.stepSize * differentiate
                differentiate = np.dot(X.T[j], np.subtract(np.dot(X,w),y))
                iteration += 1
            differentiateList[j] = abs(differentiate)
            postLoss = self.lossFunction(w,X,y)
            lossChangeList[j] = abs(postLoss - preLoss)            
            lossList.append(postLoss)
            wUpdatingPath.append(w.copy())
            if 0 in lossChangeList:
                j = (j + 1)%14
            else:
                j = np.argmax(lossChangeList)
        self.loss.append(lossList)
        self.wUpdatingList.append(wUpdatingPath)
        return w


    def generateCoeffRandom(self, X,y):
        w = np.zeros(14)
        j = random.randint(0,13)
        iteration = 0
        differentiate = np.dot(X.T[j], np.subtract(np.dot(X,w),y))
        lossList = []
        differentiateList = np.zeros(14)
        wUpdatingPath = []
        while iteration <= self.maxIter and (j == 13 or 0 in differentiateList or max(differentiateList) > self.converge):
            differentiateList[j] = differentiate
            w[j] -= self.stepSize/self.numTrainingData * differentiate
            j = random.randint(0,13)
            differentiate = np.dot(X.T[j], np.subtract(np.dot(X,w),y))
            iteration += 1
            if iteration % 1e2 == 0:
                lossList.append(self.lossFunction(w,X,y))
                wUpdatingPath.append(w.copy())
        self.loss.append(lossList)
        self.wUpdatingList.append(wUpdatingPath)
        return w
    
    def predict(self, w,test):
        poss = np.dot(w,test)
        return np.argmax(poss)+1 
    
    def testErrorPath(self, testData, testLabel):
        errorPath = []
        maxLen = max(len(self.wUpdatingList[0]),len(self.wUpdatingList[1]),len(self.wUpdatingList[2]))
        self.wUpdatingList[0] += self.wUpdatingList[0][-1]*(maxLen-len(self.wUpdatingList[0]))
        self.wUpdatingList[1] += self.wUpdatingList[1][-1]*(maxLen-len(self.wUpdatingList[1]))
        self.wUpdatingList[2] += self.wUpdatingList[2][-1]*(maxLen-len(self.wUpdatingList[2]))

        for i in range(0,len(self.wUpdatingList[0])):
            curW = [self.wUpdatingList[0][i],self.wUpdatingList[1][i],self.wUpdatingList[2][i]]
            errorPath.append(self.testSpecific(curW, testData, testLabel))
        return errorPath
    
    def testSpecific(self, w, testData, testLabel):
        error = 0
        for i in range(0,len(testData)):
            if predict(w, testData[i]) != testLabel[i]:
                error += 1
        return error*1.0/len(testData)
    
    def test(self, testData, testLabel):
        error = 0
        for i in range(0,len(testData)):
            if predict(self.coeffi, testData[i]) != testLabel[i]:
                error += 1
        return error*1.0/len(testData)
```

```python
def plotUpdate(batch, first, second, title = 'L(w1,t) decreases with t', ylabel = 'L(w1)'):
    inOrder, = plt.plot([i*batch for i in range(0,len(first))], first)
    random, = plt.plot([i*batch for i in range(0,len(second))], second)
    plt.xlabel('t')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend([inOrder, random], ['in order-feature coordinate descent', 'random-feature coordinate descent'])
    plt.grid(True)
```

```python
class CoordinateDescentMulti():
    def __init__(self):
        self.stepSize = 1e-6
        self.maxIter = 1e6
        self.converge = 1e-5
        self.coeffi = []
        self.numClass = 3
        self.trainingData = None
        self.trainingLabel = None
        self.numTrainingData = None
        self.loss = []
        self.wUpdatingList = []
        self.Y = None
        
    def lossFunction(self,w,trainData,trainLabel):

        lossList = map(lambda x: np.log(1+np.exp(-x[-1]*(np.dot(x[:-1],w)))), zip(trainData,trainLabel))
        loss = reduce(lambda x,y: x+y, lossList)
        return loss
    
    def libLoss(self, w1,w2,w3):
        def pred(coeff,data):
            return np.exp(np.dot(coeff,data))  
        predictProbability = map(lambda x: [pred(w1,x)/(pred(w1,x)+pred(w2,x)+pred(w3,x)),
                        pred(w2,x)/(pred(w1,x)+pred(w2,x)+pred(w3,x)),
                        pred(w3,x)/(pred(w1,x)+pred(w2,x)+pred(w3,x))], trainingData)        
        return metrics.log_loss(trainingLabel,predictProbability)
    def myLossFuction(self,w1,w2,w3):
        return self.lossFunction(w1,self.trainingData,self.Y[0])+self.lossFunction(w2,self.trainingData,self.Y[1])+self.lossFunction(w3,self.trainingData,self.Y[2])
    
    def fit(self, trainingData, trainingLabel, randomize = False):
        self.trainingData = trainingData
        self.trainingLabel = trainingLabel
        self.numTrainingData = len(trainingLabel)
        self.Y = -np.ones((3,len(trainingData)))
        for i in range(0,len(trainingLabel)):
            self.Y[int(trainingLabel[i])-1][i] = 1
            
        for curClass in range(self.numClass):
            if randomize:
                curCoeff = self.generateCoeffRandom(trainingData, self.Y[curClass])
            else:
                curCoeff = self.generateCoeffBatch(trainingData, self.Y[curClass])
            self.coeffi.append(curCoeff)
    
    def generateCoeffBatch(self, X,y):
        w = np.zeros(14)
        j = 0
        iteration = 0
        lossList = []
        wUpdatingPath = []
        lossChangeList = np.zeros(14)
        differentiateList = np.zeros(14)
        differentiate = np.dot(X.T[j], np.subtract(1/(1+np.exp(np.dot(X,w))),y))
        postLoss = self.lossFunction(w,X,y)
        while iteration <= self.maxIter:
            preLoss = postLoss
            for t in range(0,500):
                w[j] -= self.stepSize * differentiate
                differentiate = -np.dot(X.T[j]*y, 1/(1+np.exp((np.dot(X,w))*y)))
                iteration += 1
            differentiateList[j] = abs(differentiate)
            postLoss = self.lossFunction(w,X,y)
            lossChangeList[j] = preLoss - postLoss         
            lossList.append(postLoss)
            wUpdatingPath.append(w.copy())
            if 0 in lossChangeList:
                j = (j + 1)%14
            else:
                j = np.argmax(lossChangeList)
            if iteration% 1e5 == 0:
                print iteration,
        self.loss.append(lossList)
        self.wUpdatingList.append(wUpdatingPath)
        return w
    
    def generateCoeffRandom(self, X,y):
        w = np.zeros(14)
        j = random.randint(0,13)
        iteration = 0
        lossList = []
        wUpdatingPath = []
        differentiateList = np.zeros(14)
        differentiate = np.dot(X.T[j], np.subtract(1/(1+np.exp(np.dot(X,w))),y))
        postLoss = self.lossFunction(w,X,y)
        while iteration <= self.maxIter and (j == 13 or 0 in differentiateList or max(differentiateList) > self.converge):
            preLoss = postLoss
            for t in range(0,500):
                w[j] -= self.stepSize * differentiate
                differentiate = -np.dot(X.T[j]*y, 1/(1+np.exp((np.dot(X,w))*y)))
                iteration += 1

            differentiateList[j] = abs(differentiate)
            postLoss = self.lossFunction(w,X,y)
            lossList.append(postLoss)
            wUpdatingPath.append(w.copy())
            
            j = random.randint(0,13)
            differentiate = np.dot(X.T[j], np.subtract(1/(1+np.exp(np.dot(X,w))),y))
            
            if iteration% 1e5 == 0:
                print iteration,
                
        self.loss.append(lossList)
        self.wUpdatingList.append(wUpdatingPath)
        return w
    
    def predict(self, w,test):
        poss = np.dot(w,test)
        return np.argmax(poss)+1 
    
    def testLossPath(self):
        lossPath = []
        for i in range(0,len(self.wUpdatingList[0])):
            lossPath.append(self.libLoss(self.wUpdatingList[0][i],self.wUpdatingList[1][i],self.wUpdatingList[2][i]))
        return lossPath
    
    def testLossPathUsingMyLossFunction(self):
        lossPath = []
        for i in range(0,len(self.wUpdatingList[0])):
            lossPath.append(self.myLossFuction(self.wUpdatingList[0][i],self.wUpdatingList[1][i],self.wUpdatingList[2][i]))
        return lossPath
    
    def testErrorPath(self, testData, testLabel):
        errorPath = []
        for i in range(0,len(self.wUpdatingList[0])):
            curW = [self.wUpdatingList[0][i],self.wUpdatingList[1][i],self.wUpdatingList[2][i]]
            errorPath.append(self.testSpecific(curW, testData, testLabel))
        return errorPath
    
    def testSpecific(self, w, testData, testLabel):
        error = 0
        for i in range(0,len(testData)):
            if self.predict(w, testData[i]) != int(testLabel[i]):
                error += 1
        return error*1.0/len(testData)
    
    def test(self, testData, testLabel):
        error = 0
        for i in range(0,len(testData)):
            if self.predict(self.coeffi, testData[i]) != int(testLabel[i]):
                error += 1
        return error*1.0/len(testData)
```
